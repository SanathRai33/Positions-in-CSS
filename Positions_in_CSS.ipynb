2. Task 2:
Problem Statement
Explain the difference between Absolute and Relative positioning.

Ans -->

1. Relative Positioning (position: relative;)
* The element is positioned relative to its original position in the normal document flow.
* Moving it with top, left, right, or bottom does not affect other elements.
* The original space is preserved, meaning no overlapping unless manually adjusted.

Example:
.relative-box {
    position: relative;
    top: 20px;
    left: 30px;
    background-color: lightblue;
}
* The element moves 20px down and 30px right from its original place.


2. Absolute Positioning (position: absolute;)
* The element is positioned relative to its nearest positioned ancestor (relative, absolute, or fixed parent).

* If no such ancestor exists, it is positioned relative to the <html> (viewport).

* It is removed from the normal document flow, meaning other elements ignore its space.

Example:
.absolute-box {
    position: absolute;
    top: 30px;
    left: 50px;
    background-color: lightcoral;
}
* The element moves 30px down and 50px right, ignoring its original position.

5. Task 5:
Problem Statement
Explain the z-index, with a code example.

Ans -->

<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Z-Index Example</title>
    <style>
        .container {
            position: relative;
            width: 500px;
            height: 270px;
            border: 2px solid black;
        }
        .box {
            position: absolute;
            width: 150px;
            height: 150px;
            color: white;
            font-size: 18px;
            text-align: center;
            line-height: 150px;
        }
        .box1 {
            background-color: red;
            top: 70px;
            left: 70px;
            z-index: 1;
            opacity: 0.9;
        }
        .box2 {
            background-color: blue;
            top: 110px;
            left: 110px;
            z-index: 2; /* Higher z-index means it appears on top */
            opacity: 0.4;
        }

        .box3 {
            background-color: green;
            top: 30px;
            left: 30px;
            z-index: 0; /* Lower z-index means it appears below others */
        }
    </style>
</head>
<body>
    <h2>Understanding Z-Index</h2>
    <div class="container">
        <div class="box box1">Z-Index: 1</div>
        <div class="box box2">Z-Index: 2</div>
        <div class="box box3">Z-Index: 0</div>
    </div>
</body>
</html>
